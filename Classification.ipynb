# Импорт и настройка библиотек для ML

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier, StackingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, train_test_split, StratifiedKFold
from sklearn.metrics import roc_auc_score
import numpy as np
```

```python
dec = DecisionTreeClassifier(random_state=42)
forest = RandomForestClassifier(random_state=42)
lg = LogisticRegression(max_iter=1000, random_state=42)
knn = KNeighborsClassifier()
naiv = GaussianNB()

boost = GradientBoostingClassifier(random_state=42)

dec_params = {'criterion': ['gini', 'entropy'], 'splitter' : ['best', 'random']}
forest_param = {'n_estimators' : np.arange(10, 110, 10), 'criterion' : ['gini', 'entropy']}
lg_param = {'C': np.arange(0.5, 1, 0.2)}
knn_param = {'n_neighbors' : np.arange(5,10,1), 'weights': ['uniform', 'distance']}
naiv_param = {'var_smoothing' : np.arange(0.000000001, 0.000001, 0.00005)}
gbc_params = {'learning_rate': np.arange(0.1, 0.6, 0.1)}

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
gs_dec = GridSearchCV(dec, dec_params, cv = skf)
gs_for = GridSearchCV(forest, forest_param, cv = skf)
gs_lg = GridSearchCV(lg, lg_param, cv = skf)
gs_knn = GridSearchCV(knn, knn_param, cv = skf)
gs_naiv = GridSearchCV(naiv, naiv_param, cv = skf)

meta = GridSearchCV(boost, gbc_params, cv = skf)
```

# Определение функции для всех моделей

```python
def all_models(x, y, models, names_model, random_state = 42, test_size = 0.2):
    x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=random_state, test_size=test_size)
    for model in models:
        model = model.fit(x_train, y_train)
    
    for step in range(len(models)):
        y_pred = models[step].predict_proba(x_test)
        prec = roc_auc_score(y_test, y_pred[:,1])
        print(f'AUC модели {names_model[step]} : \n{prec}')
        print('\t' + str(models[step].best_params_))
```

```python
y = data['is_loyality']
x = data.loc[:, data.columns != 'is_loyality']
models_ = [gs_dec, gs_for, gs_lg, gs_knn, gs_naiv]
names_ = ['Деревья решений', 'Случайный лес', 'Логистическая регрессия', 'Ближайшие соседи', 'Наивный байес']
```

```python
all_models(x, y, models = models_, names_model=names_, random_state=42, test_size=0.2)
```

```python
estimators = [
    ('dec',  DecisionTreeClassifier(random_state=42, criterion='entropy', splitter='best')),
    ('forest', RandomForestClassifier(random_state=42, criterion='entropy', n_estimators = 60)),
    ('lg', LogisticRegression(max_iter=1000, random_state=42, C=0.8999999999999999)),
    ('knn', KNeighborsClassifier(n_neighbors=9, weights='uniform')),
    ('naiv', GaussianNB(var_smoothing=1e-09))
]
final_estimator = GradientBoostingClassifier(n_estimators=25, random_state=42)
reg = StackingClassifier(estimators=estimators, final_estimator=final_estimator)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=17, test_size=0.2)
```

## Проход по всем моделям

```python
all_models(x, y, models = models_, names_model=names_, random_state=42, test_size=0.2)
```

## Стекинг

```python
mod = reg.fit(x_train, y_train)
y_pred = mod.predict_proba(x_test)
auc = roc_auc_score(y_test, y_pred[:,1])
print(f'Stakcing auc: {auc}')
```
